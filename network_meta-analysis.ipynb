```r
#* Apply meta function
m.netmeta <- netmeta(TE = TE,
    seTE = seTE,
    treat1 = treat1,
    treat2 = treat2,
    studlab = author,
    data = TherapyFormats,
    sm = "SMD",
    comb.fixed = TRUE,
    comb.random = FALSE,
    reference.group = "cau",
    details.chkmultiarm = TRUE,
    sep.trts = " vs"
)
```

```r
m.netmeta
```

```r
decomp.design(m.netmeta)
```

```r
#* studies labels
m.netmeta$trts
```

```r
#* exchange short labels for long labels
long.labels <- c("Care AsUsual", "Group",
    "Guided Self-Help",
    "Individual", "Telephone",
    "Unguided Self-Help",
    "Waitlist"
)
#* plot netgraph
netgraph(
    m.netmeta,
    labels = long.labels
)
```

```r
#* plot 3d graph
#* It's possible some errors to renderer the 3D plot inside jupyter
netgraph(
    m.netmeta,
    labels = long.labels,
    dim = "3d"
)
```

```r
direct.evidence.plot(m.netmeta)
```

```r
result.matrix <- m.netmeta$TE.fixed
result.matrix <- round(result.matrix, 2)
result.matrix[lower.tri(result.matrix, diag = FALSE)] <- NA
result.matrix
```

```r
netleague(
    m.netmeta,
    bracket = "(",
    digits=2
)
```

```r
forest(
    m.netmeta,
    reference.group ="cau",
    sortvar = TE,
    xlim =c(-1.3, 0.5),
    smlab =paste("Therapy Formatsvs.CareAsUsual\n",
    "(Depressive Symptoms)"),
    drop.reference.group = TRUE,
    label.left ="FavorsIntervention",
    label.right ="FavorsCareAsUsual",
    labels = long.labels
)
```
